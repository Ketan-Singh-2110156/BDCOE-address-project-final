```python
def check_pin(x):
    url="https://api.openweathermap.org/geo/1.0/zip?zip="+str(x)+",US&appid=e0bd39d1bcd127c4b83811ddc55c23f2"
    api_link = requests.get(url)
    api_data = api_link.json()
    try:
        return api_data['name']
    except:
        return "none"
```

```python
us['Mailing Zip Code']=us.loc[:,'Mailing Zip Code'].apply(lambda x: math.floor(x))
us
```

```python
def check_pin(pin): #returns city for nigerian contries
    url="https://city-pin.onrender.com/api/v1/city?country=india&pincode="+str(pin)
    try:
        api_link = requests.get(url)
        api_data = api_link.json()
        return api_data["city"]
    except:
        return "none"
```

```python
def get_city(x):
    url="https://city-pin.onrender.com/api/v1/pincode?country=india&city="+str(x).lower()
    try:
        api_link = requests.get(url)
        api_data = api_link.json()
        return api_data["pincode"][0]
    except:
        return "none"
```
