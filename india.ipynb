```python
def getpin(city):
    url="https://api.postalpincode.in/postoffice/"+city
    try:
        api_link = requests.get(url)
        api_data = api_link.json()
        return api_data[0]['PostOffice'][0]['Pincode']
    except:
        return "none"
```
