```python
def getpin(city):
    url="https://api.postalpincode.in/postoffice/"+city
    api_link = requests.get(url)
    api_data = api_link.json()
    try:
        return api_data[0]['PostOffice'][0]['Pincode']
    except:
        return "none"
```

```python
india_nullpin = india[india['Postal Code'].isnull()]
india_nullpin
```

```python
india_nullpin['Postal Code']=india_nullpin.loc[:,'City'].apply(lambda x: getpin(x))
```
