```python
def getpin(city):
    url="https://api.postalpincode.in/postoffice/"+city
    try:
        api_link = requests.get(url)
        api_data = api_link.json()
        return api_data[0]['PostOffice'][0]['Pincode']
    except:
        return "none"
```

```python
india_nullpin = india[india['ZIP/Postal Code'].isnull()]
india_nullpin['City'].isnull().sum() #there are total 8 values where city and zip code both are not given
```

```python
india_nullpin
```

```python
india.dropna(subset="ZIP/Postal Code",inplace=True)
```

```python
india_invalid = india_nullpin[india_nullpin['City'].isnull()]
india_nullpin.dropna(subset="City",inplace=True)
```

```python
india_nullpin['ZIP/Postal Code']=india_nullpin.loc[:,'City'].apply(lambda x: getpin(x))
```
